```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.svm import SVC
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
```

```python
df=pd.read_csv(r'D:\Data_Science\7-Machine_Learning\projects\END-TO-END_projectsML\parkinsons_END_TO_END_ML\NoteBook\parkinsons_after_edit.csv')
```

```python
df.head()
```

```python
x=df.drop('status',axis=1)
y=df['status']
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.1, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.metrics import f1_score, classification_report, recall_score


def evaluate(y_test,y_pred):
    f1=f1_score(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    return f1,report,recall
```

```python
models={
    "LogisticRegression":LogisticRegression(),
    "DecisionTree":DecisionTreeClassifier(),
    "KNN":KNeighborsClassifier(),
    "RandomForest":RandomForestClassifier(),
    "AdaBoost":AdaBoostClassifier(),
    "SVC":SVC(),
    "CatBoost":CatBoostClassifier(),
    "XGBoost":XGBClassifier()
}
```

```python
model_list_name=[]
evaluate_model=[]

for name,model in models.items():
    #train
    model.fit(x_train,y_train)
    #predict
    predict_train=model.predict(x_train)
    predict_test=model.predict(x_test)
    #evaluate
    f1_train,report_train,recall_train=evaluate(y_train,predict_train)
    f1_test,report_test,recall_test=evaluate(y_test,predict_test)

    print(name)
    model_list_name.append(name)

    print("f1_train",f1_train)
    print("report_train",report_train)
    print("recall_train",recall_train)
    print("---------------------------------------------------")
    print("f1_test",f1_test)
    print("report_test",report_test)
    print("recall_test",recall_test)

    print("="*35)
    print('\n')

    evaluate_model.append(f1_test)
```

```python
result_df=pd.DataFrame({
    "modelName":model_list_name,
    "Evaluate_model":evaluate_model
})
result_df.head()
```

```python
best_model=result_df.iloc[0]['modelName']
best_accuracy=result_df.iloc[0]['Evaluate_model']


print("The Best Model Call Name = ",best_model)
print("The High score Or High accuracy = ",best_accuracy)
```
